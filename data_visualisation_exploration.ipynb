Data Visualization

```python
# Libaries needed for visualisation.
%matplotlib inline
import pandas as pd
# The Matplotlib and Seaborn Python Libraries is used to create simple visualizations.
import matplotlib.pyplot as plt
import seaborn as sns 
sns.set()
```

```python
df = pd.read_csv("cleaned_data_v5.csv") # Loading files 
```

```python
df.head()# First 5 tuples in dataset
```

```python
df['gender'].value_counts() # Get a count of the unique values, gender
```

```python
labels = ["Female", "Male"]
colors = ['blue','orange'] # colours 
explode = [0.06,0] # explode the first slice
fig, ax1 = plt.subplots(figsize =(10,7))
ax1.pie(df['gender'].value_counts()
        ,colors=colors,explode=explode, startangle=90, autopct='%1.1f%%')
plt.title("Diabetes Patients by Gender", fontsize= 24)
ax1.legend(labels, loc="upper right")
plt.tight_layout()
plt.show
```

```python
df['race'].value_counts()
```

```python
sns.set(font_scale=1.4)
df['race'].value_counts().plot(kind='bar', figsize=(10,6),rot=0)
plt.xlabel("Race/Ethnicity", labelpad=14)
plt.ylabel("Number of patients", labelpad=14)
plt.title("Diabetes Patients by Race/Ethinicity", y=1.06);
```

```python
df['age'].value_counts()
```

```python
df['age'].value_counts().sort_index().plot(kind='barh', figsize=(8,6))
plt.xlabel("Number of Patients", labelpad=14)
plt.ylabel("Age",labelpad=14)
plt.title("Number of Patients with Diabetes by Age",y=1.02);
```

```python
df['readmitted'].value_counts()# Get a count of the unique values
```

```python
labels = ["Have not been readmitted", "Readmitted after 30 days or more", "Readmitted within 30 Days"]
colors = ['pink','lightblue','lightgreen'] # colours 
explode = [0.09,0,0] # explode the third slice
fig, ax1 = plt.subplots(figsize =(17,7))
ax1.pie(df['readmitted'].value_counts()
        ,colors=colors,explode=explode, startangle=90, autopct='%1.1f%%')
plt.title("Patients Readmittance", fontsize= 24)
ax1.legend(labels, loc="upper right")
plt.tight_layout()
plt.show
```

```python
df['days_in_hospital'].value_counts()# Get a count of the unique values, days in hospital
```

```python
sns.set(font_scale=1.4)
df['days_in_hospital'].value_counts().sort_index().plot(kind='bar', figsize=(7,6), rot=0)
plt.xlabel("Days in Hospital", labelpad=14)
plt.ylabel("Number of Patients",labelpad=14)
plt.title("Days spent in Hopsital");
```

```python
df['change_in_meds'].value_counts()# Get a count of the unique values, change in medication
```

```python
sns.set(font_scale=1.4)
df['gender'].groupby(df['change_in_meds']).value_counts().plot(kind='bar', figsize=(7,6), rot=0)
plt.xlabel("Changes", labelpad=14)
plt.ylabel("Number of Patients",labelpad=14)
plt.title("Change in Medication");
```

```python
df['num_current_medications'].value_counts(bins=10)
```

```python
df['num_current_medications'].value_counts(bins=10).sort_index().plot(kind ='bar', 
                              figsize=(18,6), rot=0)
plt.xlabel("Number of Medication", labelpad=14)
plt.ylabel("Number of Patients",labelpad=14)
plt.title("Number of Medication used by Patients");
#bins reduce numbers 
```

```python
df['diabates_med_prescribed'].value_counts()
```

```python
labels = ["Yes", "No"]
colors = ['orange','lightgrey'] # colours 
explode = [0,0.09] # explode the second slice
fig, ax1 = plt.subplots(figsize =(20,6))
ax1.pie(df['diabates_med_prescribed'].value_counts()
        ,colors=colors,explode=explode, startangle=90, autopct='%1.1f%%')
plt.title("Have patients been prescribed Diabetes Medication?", fontsize= 20)
ax1.legend(labels, loc="upper right")
plt.tight_layout()
plt.show
```
